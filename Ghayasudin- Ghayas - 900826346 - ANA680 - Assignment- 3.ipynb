```python
outlier_indices = []
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_idx = data[(data[col] < lower) | (data[col] > upper)].index
    outlier_indices.extend(outlier_idx)

# Unique outlier indices
outlier_indices = list(set(outlier_indices))
print(f"Total unique rows with any outlier: {len(outlier_indices)}")
```

2-  Cap or Winsorize (Limit Values)

Best if:
- We want to keep all rows but limit the influence of extreme values.
Result :
- Keeps the dataset size same
- Prevents extreme values from skewing the model
- May slightly distort true distributions

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data[col] = np.where(data[col] < lower, lower, data[col])
    data[col] = np.where(data[col] > upper, upper, data[col])

Best if

###   c. Linear SVM (kernel = linear)

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

#  Create Linear SVM model
linear_svm = SVC(kernel='linear', probability=True, random_state=42)

# Train on scaled data
linear_svm.fit(X_train_scaled, y_train.ravel())

# Make predictions
y_pred_linear_svm = linear_svm.predict(X_test_scaled)

# Evaluate performance
print("=== Linear SVM (Kernel = Linear) ===")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_linear_svm))
print("\nClassification Report:\n", classification_report(y_test, y_pred_linear_svm))

# ROC-AUC
y_proba_linear_svm = linear_svm.predict_proba(X_test_scaled)[:, 1]
roc_auc = roc_auc_score(y_test, y_proba_linear_svm)
print("ROC-AUC Score:", roc_auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba_linear_svm)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Linear SVM")
plt.legend()
plt.show()
```
